```python
# Import Numpy for calculations, matplotlib for charting, and pandas for coding
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Save path to data set in a variable
data_file = "../resources/raw/dataSet.csv"
data_file2 = "../resources/raw/counties.csv"
```

```python
# Use Pandas to read data
data_file_df = pd.read_csv(data_file)
data_file_df.head()
```

```python
#Drop NaN's from columns

data_file_df.dropna(inplace=True, subset=["sub_region_1", "sub_region_2"])
data_file_df.head(20)
```

```python
data_file2_df = pd.read_csv(data_file2)
data_file2_df.head() 
```

```python
combined_df = data_file_df[data_file_df["sub_region_1"].isin(data_file2_df["State"])]
combined_df = combined_df[combined_df["sub_region_2"].isin(data_file2_df["County"])]
#combined_df.loc[combined_df["sub_region_2"]=="Sacramento County",:]
combined_df.head()
#combined_df["sub_region_1"].value_counts()
#data_file2_df["County"].tolist()
```

```python
combined_df.to_csv("../resources/clean/Karen_clean.csv")
```

```python
#load the csv
whole_df = pd.read_csv("../resources/clean/Karen_clean.csv")
whole_df
```

```python
#make the date column a datetime object
whole_df['date'] = pd.to_datetime(whole_df['date'], format="%Y/%m/%d")
```

```python
#add a new column that is the month
whole_df["month"] = whole_df["date"].map(lambda x: x.month)
#group on regions and month
grouped_df = whole_df.groupby(["sub_region_1","sub_region_2","month"]).mean()
grouped_df = grouped_df[["retail_and_recreation_percent_change_from_baseline","grocery_and_pharmacy_percent_change_from_baseline","parks_percent_change_from_baseline","transit_stations_percent_change_from_baseline","workplaces_percent_change_from_baseline","residential_percent_change_from_baseline"]]
grouped_df
```
